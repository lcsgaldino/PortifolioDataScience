# Projeto de Machine Learning

**Proposta:** Treinar o modelo para prever tipos de Vinho, se ele é tinto ou branco.

```python
#Biblioteca para tratar os dados, importar o csv
import pandas as pd

#Biblioteca especifica p/ machine learning, aqui criamos os conjuntos de dados de treino e teste
from sklearn.model_selection import train_test_split

#Biblioteca pra criação do modelo
from sklearn.ensemble import ExtraTreesClassifier
```

```python
df = pd.read_csv('wine_dataset.csv')
```

```python
df.head()
```

```python
#Alterando a variavel red para o número inteiro 0
df['style'] = df['style'].replace('red', 0)
```

```python
#Alterando a variavel white para o número inteiro 1
df['style'] = df['style'].replace('white', 1)
```

```python
#separando as variáveis entre preditores e váriaveis alvo
y = df['style']
x= df.drop('style',axis =1)
```

```python
#criando conjuntos de daos de treino e teste
x_treino, x_teste, y_treino , y_teste =train_test_split(x,y, test_size =0.3)
```

```python
y.head
```

**ExtraTreesClassifier é um algoritimo de ML que ira usar os dados de treino acima definodos**

```python
from sklearn.ensemble import ExtraTreesClassifier
#Criando o Modelo
modelo =ExtraTreesClassifier()
modelo.fit(x_treino,y_treino)

#imprimindo resultado:
resultado =modelo.score(x_teste,y_teste)
print('Acurácia:',resultado)
```
